```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
plt.plot(arr)
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4,5]
y= [2,3,5,7,11]
plt.scatter(x,y,c='r')
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()
```

```python
height = [10,20,30,40,50]
age = [20,25,30,35,40]
plt.bar(height,age)
plt.xlabel("Height")
plt.ylabel("Age")
plt.title("Bar Chart")
plt.show()
```

```python
plt.hist(height)
```

```python
data =[1,2,2,2,4,5,5,6,5,3,4,1,2,2,2,3]
plt.hist(data,bins=5)
plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C','Category D']
values = [10, 20, 15, 25]
explode = (0.1, 0, 0, 0)
plt.pie(values,labels=categories,autopct='%1.1f%%',explode=explode)
plt.show()
```
