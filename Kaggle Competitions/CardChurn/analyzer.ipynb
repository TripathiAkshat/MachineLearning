```python
import pandas as pd
```

```python
testdf = pd.read_csv('test.csv')
traindf = pd.read_csv('train.csv')
```

```python
traindf = traindf.iloc[0:10000]
```

```python
traindf.drop(columns=['id','CustomerId','Surname'],inplace = True)
```

```python
traindf = pd.get_dummies(traindf,columns=['Geography'])
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
traindf['Gender'] = le1.fit_transform(traindf['Gender'])
```

```python
traindf
```

```python
X = traindf[['CreditScore','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary','Geography_France','Geography_Spain']]
```

```python
Y = traindf['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
from sklearn.preprocessing import StandardScaler
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
y_pred = clf.predict(X_test)
print(confusion_matrix(Y_test,y_pred))
print("The accuracy is ",accuracy_score(Y_test,y_pred))
```
